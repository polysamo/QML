```python
import random
from quantumnet.components import Network, Controller
from quantumnet.objects import Logger

def validate_clients_and_server_in_graph(graph, clients, server):
    missing_nodes = [node for node in clients + [server] if node not in graph]
    if missing_nodes:
        raise ValueError(f"Nós ausentes na topologia: {missing_nodes}")

def run_simulation_with_slices(protocols, clients, num_requests, simulation_id):
    rede = Network()
    controlador = Controller(rede)

    # Configurar a topologia
    server = 10
    rede.set_topology_for_slices(graph_type='grade', dimensions=(4, 4), clients=clients, server=server)
    validate_clients_and_server_in_graph(rede._graph, clients, server)

    slice_paths_list = rede.run_slice_simulation(clients=clients, server=server)
    
    # Inicializar os slices
    controlador.initialize_slices(
        network=rede,
        clients=clients,
        server=server,
        protocols=protocols,
        slice_paths_list=slice_paths_list
    )

    # Gerar requisições
    requests = []
    for i in range(num_requests):
        client_index = i % len(clients)
        alice_id = clients[client_index]
        protocol = protocols[client_index]
        slice_path = slice_paths_list[client_index]
        bob_id = server
        num_qubits = 10
        num_gates = 20

        request = rede.generate_request_slice(
            alice_id=alice_id,
            bob_id=bob_id,
            num_qubits=num_qubits,
            num_gates=num_gates,
            protocol=protocol,
            slice_path=slice_path,
            scenario=2
        )
        requests.append(request)

    scheduled_timeslots = controlador.schedule_requests(
        requests=requests,
        slice_paths=controlador.slices,
        protocols=protocols
    )

    rede.execute_scheduled_requests(scheduled_timeslots, slice_paths=controlador.slices)
    report = controlador.print_report(scheduled_timeslots, slice_paths=controlador.slices)
    total_used_eprs = rede.get_total_useds_eprs()
    avg_fidelity = rede.application_layer.avg_fidelity_on_applicationlayer()

    return {
        "simulation_id": simulation_id,
        "protocols": protocols,
        "success_count": report["success_count"],
        "failure_count": report["failure_count"],
        "total_eprs_used": total_used_eprs,
        "average_fidelity": avg_fidelity,
    }

def run_multiple_simulations(num_simulations, num_requests, protocols, clients, output_file):
    import csv

    with open(output_file, mode='w', newline='') as csvfile:
        fieldnames = ["simulation_id", "protocols", "success_count", "failure_count", "total_eprs_used", "average_fidelity"]
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()

        for i in range(num_simulations):
            print(f"Running simulation {i+1}/{num_simulations}...")
            metrics = run_simulation_with_slices(protocols, clients, num_requests, i+1)
            writer.writerow(metrics)
            print(f"Simulation {i+1} completed: {metrics}")
#[13, 8, 2, 6, 0, 3, 11, 15]
def main():
    protocols = ["AC_BQC", "BFK_BQC","AC_BQC","BFK_BQC","AC_BQC","BFK_BQC","AC_BQC","BFK_BQC"] # Escolha do protocolo 
    clients = [13, 8, 2,6,0,3,11,15] # Escolha dos clientes
    run_multiple_simulations(
        num_simulations=10,
        num_requests=50,
        protocols=protocols,
        clients=clients,
        output_file="simulation2_results.csv"
    )

if __name__ == "__main__":
    main()
```
